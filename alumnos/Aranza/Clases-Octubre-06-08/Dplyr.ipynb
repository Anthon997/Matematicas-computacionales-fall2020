## Sumarize

```r
#Calcula el retrazo promedio y quítame el NA
summarise(flights,
delay = mean(dep_delay, na.rm = TRUE))
```

## Agrupaciones
### Group by

```r
flights %>% 
    group_by(month, day) %>%
        summarise(delay = mean(dep_delay, na.rm = TRUE)) %>% 
            arrange(desc(delay))
#Agrupamos --> funccion para delay --> ordenamos descendente
#Así tenemos el delay por mes y dia
#El peor día para viajar sería el 08 de marzo
```

## Preguntas

    ¿Hay algún día de la semana que sea considerablemente mejor para volar?

```r
library(lubridate)
```

```r
flights %>%
mutate(
week_day = wday(mdy(paste0(month,'-',day,'-',year))))%>% 
group_by(week_day) %>%
summarise(delay = mean(dep_delay, na.rm = TRUE)) 
```

    ¿Hay alguna aerolínea que tenga algún problema a nivel mes?

```r
flights %>%
    group_by(carrier, month) %>%
        summarise(delay = mean(dep_delay, na.rm =TRUE)) %>%
            arrange(desc(delay))
```

    ¿Hay algún avión problemático?

```r
flights %>%
    group_by(tailnum) %>%
        summarise(delay = mean(dep_delay, na.rm = TRUE)) %>%
            arrange(desc(delay))
```

    ¿Hay algún destino/origen problemático?

```r
flights %>%
group_by(dest) %>%
summarise(delay = mean(dep_delay,na.rm = TRUE)) %>%
arrange(desc(delay))
```

    ¿Hay alguna correlación entre distancia y retrasos?

```r
flights %>%
    group_by(tailnum, distance) %>%
        summarise(delay = mean(dep_delay, na.rm = TRUE)) %>%
            arrange(desc(delay))
```
