```python
%pylab inline
import numpy as np 
import matplotlib.pyplot as plt
from sympy import Symbol 
```

```python
x = np.linspace(-1, 1.)
cos = np.cos(x)
X = cos*x
X2 = cos*(x**2)
X3 = cos*(x**3)
X4 = cos*(x**4)
```

```python
plt.figure(figsize=(8, 6), dpi=80)

plt.plot(x,X, color = "red", label="x*cos(x)")
plt.plot(x, X2, color = "green", label="(x^2)*cos(x)")
plt.plot(x, X3, color = "brown", label="(x^3)*cos(x)")
plt.plot(x, X4, color = "blue", label="(x^4)*cos(x)")
plt.legend(loc='lower right')
plt.title("Gr√°fica de potencias de x multiplicadas por coseno\n")


plt.xlim(x.min() * 1.1, x.max() * 1.1)


ax = plt.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none')   # El borde de la gr√°fica sin color. Borde derecho 
ax.spines['top'].set_color('none')     # El borde de la gr√°fica sin color. Borde superior
ax.xaxis.set_ticks_position('bottom')      #Etiquetas del eje x debajo
ax.spines['bottom'].set_position(('data',0))   #Posicionas los datos alrededor de cero
ax.yaxis.set_ticks_position('left')      #Etiquetas del eje y a la izquierda
ax.spines['left'].set_position(('data',0))   #Centras los datos alrededor de cero 
```

```python
fig, ax = plt.subplots(2,2, figsize=(10,10))   #La primera pareja ordenada te dice como van a estar organizados los subplots
ax[0,0].plot(x,X, color = "red", label="x*cos(x)")

ax[0,1].plot(x,X2, color = "green", label="(x^2)*cos(x)")

ax[1,0].plot(x,X3, color = "brown", label="(x^3)*cos(x)")

ax[1,1].plot(x,X4, color = "black", label="(x^4)*cos(x)")


```
