# Proyecto 1- Multiples of 3 and 5

If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.

Find the sum of all the multiples of 3 or 5 below 1000.

```python
#Creamos una lista vacía para después poder agregarle los valores
lista = []
```

```python
#Checa que números abajo de 1000 son múltiplos de 3
for i in range (1000):
    if i % 3 == 0:
            lista.append(i)
```

```python
#Imprime la longitud de los múltiplos de 3 menores a 1000 y la lista de todos los números 
print (len (lista))
print (lista)
```

```python
#Checa que números abajo de 1000 son múltiplos de 5 y pega los que todavía no están en la lista 
for j in range (1000):
    if j % 5 == 0:
        if not j in lista:
            lista.append(j)
```

```python
#Imprime la longitud de los múltiplos de 5 menores a 1000 y la lista de todos los números 
print (len (lista))
print (lista)
```

```python
#Para ver si estamos bien hacemos una comprobación 
TotalReal= round((1000/3)+(1000/5)-(1000/15))
Total=len(lista)
print (TotalReal - Total)
```

```python
#Sumamos todos los números de la lista
suma = 0
for x in lista:
    suma += x
print ("La suma de los múltiplos de 3 o 5 es:",suma)
```
