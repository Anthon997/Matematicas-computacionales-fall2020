```python
import numpy as np
```

```python
#Creamos un arreglo booleano de 100 elementos, todos True
bool_arr = np.ones((100,), dtype=bool)
```

```python
#Igualamos los elementos 0 y 1 a False
bool_arr[0]=False
bool_arr[1]=False

# Para el elemento 2 en adelante, igualamos sus multiplos a False
#Esto lo hacemos con un for y con el slicing de arreglo, el for va a recorrer desde el numero 2, hasta el número para el cual
#todos sus multiplos ya estarán tachados.

#Entonces con el slicing decimos que iguale a False los elementos del arreglo a partir del
#doble del número en el cual se encuentre el ciclo for, hasta el final del arreglo, 
#en incrementos del número en el que se encuentra el ciclo for.

#De esta manera, nos aseguramos que los índices multiplos del numero al que corresponda el ciclo for
#sean igualados a False
N_max = int(np.sqrt(len(bool_arr)))
for j in range(2, N_max):
    bool_arr[2*j::j] = False
```

```python
def criba_erastotenes(maximo):
    arreglo_bool=np.ones((maximo,), dtype=bool)
    maxi=int(np.sqrt(maximo))
    for i in range(2,maxi):
        arreglo_bool[2*i:maximo:i]=False
        arreglo_bool[0]=False
        arreglo_bool[1]=False
        print(arreglo_bool)
```

```python
criba_erastotenes(10)
```

```python
for x in range(20):
    criba_erastotenes(10*x)
```
