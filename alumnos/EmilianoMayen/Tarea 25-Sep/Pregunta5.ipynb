```python
import numpy as np
%pylab inline
import matplotlib.pyplot as plt
```

```python
#Cargamos el archivo a un arreglo numpy
#Asumimos que el archivo esta guardado en el mismo directorio
arr=np.loadtxt("sunspots.txt")
```

```python
#Eliminamos los el numero de manchas solares del año 1749
arr
Y=arr[12:,1]
```

```python
sun=[x for x in Y]
sunspots=[]
```

```python
#A partir de 1750 y cada 25 años, se suma el numero de manchas solares de todo el año
# y este numero se añade a la lista 'sunspots'
suma=0
for i in range(0, len(sun),(12*25)):
    for x in range(0,12):
        suma+=sun[i+x]
    sunspots.append(suma)
    suma=0
print(sunspots)
```

```python
# Se crea un espacio lineado desde el año 1750 hasta el 2000, con un total de 11 elementos
# (250 años corresponde a 10 elementos, pero se añade el año "0", que en este caso es el año es 1750)
year=np.linspace(1750,2000,11)
```

```python
# A la lista cumsum se le va añadiendo la suma cumulativa desde el año 1750 hasta el año correspondiente
# A la lista average se le añade el "moving average" del año correspondiente
average=[]
cumsum=[]
cumsum.append(sunspots[0])
i=1
while i <len(sunspots):
    cumsum.append(sunspots[i]+cumsum[i-1])
    i+=1
for x in range(len(cumsum)):
    average.append(cumsum[x]/(x+1))
```

```python
#Graficamos
plt.figure(figsize=(8, 6), dpi=80)
plt.plot(year, sunspots, color="blue", linewidth=2.5, linestyle="-", label="Manchas Solares")
plt.plot(year, average, color="red", linewidth=2, linestyle="--", label="Promedio movil de manchas solares")
plt.xlabel("Año")
plt.ylabel("Manchas solares durante el año")
plt.legend(loc='best')
plt.grid();
plt.title('Gráfica de manchas solares cada 25 años',fontsize=20)
```
