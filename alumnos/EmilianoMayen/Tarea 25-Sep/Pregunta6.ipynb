```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Cargamos el archivo a un arreglo numpy
stars=np.loadtxt("stars.txt")
```

```python
temperatura=stars[:,0]
magnitud=stars[:,1]
```

```python
# Para obtener el logaritmo de la luminosidad o "magnitud", sus elementos deben ser estrictamente positivos,
# Por lo que nos desacemos de los que tienen valores negativos, y sus temperaturas correspondientes

temperatura=temperatura[np.where(magnitud>0)[0]]
magnitud=magnitud[np.where(magnitud>0)[0]]
```

```python
brillo=np.log10(magnitud)
```

```python
brillo
```

```python
plt.scatter(temperatura,brillo, alpha=0.3, cmap='viridis')
plt.grid();
plt.title('Diagrama de Hertzsprung-Russell para algunas estrellas de la Via Láctea',fontsize=20)
plt.ylabel("Logaritmo base 10 del brillo")
plt.xlabel("Temperatura (K)")
```
