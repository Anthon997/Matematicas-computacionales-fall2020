### Problema 1

(a) Usando `Simpy`, declara las funciones:
$$
y(x) = \sin(x)
$$

$$
z(x) = \cos(x)
$$

$$
w(x) = \frac{1}{\cos(x) + \sin(2x)}
$$

```python
#Establecimiento de símbolos y funciones a variables
y = Function('y')
z = Function('z')
w = Function('w')
x = Symbol('x')
#Fórmulas a utilizar
s = sin(x)
c = cos(x)
fun = 1/(c+sin(2*x))
```

```python
#Se da formato de escritura a la función coseno (igualación de z(x) a coseno)
coseno = Eq(z(x), c)
coseno
```

```python
#Se da formato de escritura a la función w(x)
funcion = Eq(w(x), fun)
funcion
```

$$

### Problema 2

El [**atractor de Rössler**](http://en.wikipedia.org/wiki/R%C3%B6ssler_attractor) esta descrito por el siguiente conjunto de ecuaciones:

$$
\frac{dx}{dt} = -y -z
$$

$$
\frac{dy}{dt} = x + ay
$$

$$
\frac{dz}{dt} = b + z(x-c)
$$

(a) Resuelva las ecuaciones numéricamente para 

$$
a=0.13 \quad b=0.2 \quad c=6.5 
$$

y condiciones iniciales

$$
x(0)=0 \quad y(0)=0 \quad z(0)=0 
$$

use el método de Runge-Kutta de 2do orden.

```python
#Constantes 
N = 1000
a = .13
b = .2
c = 6.5
tau = 100
dt = tau/(float)(N-1) 
tiempo = np.linspace(0, tau, N)
```

```python
#Estado inicial
i = np.zeros([N,3])
i[0,0] = 0 #x(0)
i[0,1] = 0 #y(0)
i[0,2] = 0 #z(0)
```

```python
#Método de Runge-Kutta de 2º orden (dado por el profesor)
def RK2(y, t, dt, derivadas):
    k0 = dt*derivadas(y, t)
    k1 = dt*derivadas(y + k0, t + dt)
    y_next = y + 0.5*(k0 + k1)
    
    return y_next
```

```python
#Defino Rossler para asignar las derivadas
def Rossler(estado, tiempo):
    g0 = -(estado[1])-(estado[2])
    g1 = estado[0] + (a*estado[1])
    g2 = b + (estado[2]*(estado[0]- c))
    return np.array([g0,g1,g2])
```

```python
#Función que incluye tanto Rössler como el método RK2 para ir calculando el avance a partir de la situación inicial
def rossler_rk(y, tiempo, dt): 
    for j in range(N-1):
        y[j+1] = RK2(y[j], tiempo[j], dt, Rossler)
        
    x_= np.array([y[j,0] for j in range(N)])
    y_= np.array([y[j,1] for j in range(N)])
    z_= np.array([y[j,2] for j in range(N)])
    
    return x_,y_,z_
```

```python
#Obtención de arreglos de numpy (a partir de la función rossler_rk) con los valores de x, y, z
x_, y_, z_ = rossler_rk(i, tiempo, dt)
```

(b) Muestra en una gráfica el comportamiento de las soluciones en el tiempo (i.e. grafica $x(t)$, $y(t)$ y $z(t)$).

```python
#Creamos una figura de 10x6 pulgadas
plt.figure(1, figsize=(10,6), dpi=80)
#Graficamos el comportamiento de x, y, z
plt.plot(tiempo, x_, label="x(t)")
plt.plot(tiempo, y_, label="y(t)")
plt.plot(tiempo, z_, label="z(t)")
#Añadimos título, leyendas y etiquetas de los ejes
plt.title('Atractor de Rössler 2D')
plt.xlabel('Tiempo')
plt.ylabel('x(t), y(t), z(t)')
plt.legend(loc="best")
```

(c) Muestra como se ve el **atractor de Rössler** en 3D (i.e. en el espacio).

```python
#Creamos una figura y graficamos en 3D los valores de x,y,z
fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection = '3d')
ax.plot(x_,y_,z_)
ax.set_title('Atractor de Rössler 3D')
```
