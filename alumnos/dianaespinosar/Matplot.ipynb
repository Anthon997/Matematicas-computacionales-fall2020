```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0., 5.)
y = np.cos(x)

# Una figura con cuatro subgr치ficas
fig, ax = plt.subplots(2,2, figsize=(10,10))

# Dibuja cada eje con diferentes escalas
ax[0,0].plot(x,y*x) # x*cosx
ax[1,0].plot(x,y*(x**2)) #x**2*cosx
ax[0,1].plot(x,y*(x**3)) # x**3*cosx
ax[1,1].plot(x,y*(x**4)); #x**4cosx
```

```python
X = np.linspace(0., 2.)
C = np.cos(x)
# Crea una figura de 8x6 pulgadas, con 80 puntos por pulgada (dots per inch = dpi)
plt.figure(figsize=(8, 6), dpi=80)

# Crea un subplot en una malla de 1x1.
plt.subplot(1, 1, 1)

plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cos(x)*x")
plt.plot(X, C*(X**2), color="red",  linewidth=2.5, linestyle="-", label="cos(x)*x2")
plt.plot(X, C*(X**3), color="green", linewidth=2.5, linestyle="-", label="cos(x)*x3")
plt.plot(X, C*(X**4), color="yellow",  linewidth=2.5, linestyle="-", label="cos(x)*x4")

plt.legend(loc='upper left')
```
